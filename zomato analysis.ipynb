```python
ratings=df_2.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Here we can observe that 
    when the rating is between the range 1.8 - 2.4 it is poor
    when the rating is between the range 2.5 - 3.5 it is Average
    when the rating is between the range 1.8 - 2.4 it is poor
    when the rating is between the range 3.5 - 3.9 it is Good
    when the rating is between the range 4.0 - 4.4 it is Very Good
    when the rating is between the range 4.5 - 4.9 it is Excellent

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
plt.show
```

```python
sns.barplot(x="Aggregate rating" ,y="Rating Count" ,hue='Rating color' ,data=ratings, palette=['blue','red','orange','yellow','green','darkgreen'])
plt.show
```

```python
ax = sns.countplot(x='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','darkgreen'])
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')
```

```python



# Step 1: Group by 'City' and sum the 'Votes'
city_votes = df_2.groupby('City')['Votes'].sum()

# Step 2: Sort the cities by votes in descending order
top_cities = city_votes.sort_values(ascending=False).head(5)

# Step 3: Plot a pie chart
plt.figure(figsize=(6, 6))
top_cities.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'orange', 'red', 'purple'])
plt.title('Top 5 Cities by Votes')
plt.ylabel('')  # Remove the y-axis label for clarity
plt.show()
```

```python
# Step 1: Group by 'City' and sum the 'aggregate rating'
Aggregate_Rating = df_2.groupby('Restaurant Name')['Aggregate rating'].sum()

# Step 2: Sort the cities by aggregate rating in descending order
top_restraunt = Aggregate_Rating.sort_values(ascending=False).head(7)

# Step 3: Plot a bar chart
plt.figure(figsize=(6, 6))
ax=top_restraunt.plot(kind='bar',color=['#00008B','#0047AB','#6495ED','#008080','#40B5AD','#96DED1','#9FE2BF'])

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')

plt.title('Top 7 Cities by Aggregate Rating')

plt.show()
```
